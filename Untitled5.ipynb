```python
X = ['Gender', 'Age', 'Occupation', 'City_Category',
       'Stay_In_Current_City_Years', 'Marital_Status', 'Product_Category_1',
       'Product_Category_2', 'Product_Category_3']
```

```python
Y = ['Purchase']
```

```python
cat_features = ['Gender','City_Category','Age','Stay_In_Current_City_Years']
```

```python
train, test1 = train_test_split(train, train_size = 0.99,random_state = 42)
```

```python
model = CatBoostRegressor(cat_features=cat_features,loss_function = 'RMSE',eval_metric='MAPE' ,early_stopping_rounds=200,learning_rate = 0.25,verbose = 100)
```

```python
model.fit(train[X],train[Y])
```

```python
test1['prediction'] = model.predict(test1[X])
```

```python
error(test1['Purchase'], test1['prediction'])
```
